```python
# read the dataset function
def readD(data):
    for i in data:
        D = []
        file = i + ".json" # write the name of file     
        with open(file, 'rb') as f: # read file line by line
            while True:
              line = f.readline()
              if line:
                temp=json.loads(line)
                D.append(temp)
              else:
                break


        D = pd.DataFrame(D) # transform json to dataframe
        globals()[i] = D
        f.close 
```

```python
bu=["business"]
readD(bu)
```
